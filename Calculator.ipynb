```python
def add(x, y):
    return x + y
def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y
def divide(x, y):
    return x / y
operations = {
    "1": add,
    "2": subtract,
    "3": multiply,
    "4": divide
}
print("Select operation: 1.Add 2.Subtract 3.Multiply 4.Divide")
choice = input("Enter choice: ")
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
print(operations[choice](num1, num2))
```
